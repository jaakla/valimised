```python
# lae andmed XML-idest CSV-sse
# osad failid tühjad - Tallinn näiteks, need eraldi

import csv
import xml.etree.ElementTree as ET
import os

def loadfile(fname):
    # Load the XML file
    if not os.path.isfile(fname):
        print(f'PUUDUB {parish_code}.xml')
        return
    
    tree = ET.parse(fname)
    root = tree.getroot()

    # Define the namespaces used in the XML file
    namespaces = {'ns': 'https://opendata.valimised.ee/schemas/detailed-voting-result/municipality/ep_rk/v1/'}

    # Create a list to store the extracted data
    data = []

    # Iterate over each voting result
    for votingResult in root.findall('ns:data/ns:votingResult', namespaces):
        adminUnitName = votingResult.find('ns:adminUnitName', namespaces).text
        ehakCode = votingResult.find('ns:ehakCode', namespaces).text

        # Iterate over each party
        for party in votingResult.findall('ns:voteDistributionByParties/ns:party', namespaces):
            partyName = party.find('ns:partyName', namespaces).text
            partyCode = party.find('ns:partyCode', namespaces).text

            # Iterate over each candidate
            for candidate in party.findall('ns:candidates/ns:candidate', namespaces):
                forename = candidate.find('ns:forename', namespaces).text
                surname = candidate.find('ns:surname', namespaces).text
                candidateRegNumber = candidate.find('ns:candidateRegNumber', namespaces).text

                # Iterate over each vote distribution cell
                for cell in candidate.findall('ns:votesDistributionRow/ns:votesDistributionCell', namespaces):
                    name = cell.find('ns:name', namespaces).text
                    value = cell.find('ns:value', namespaces).text

                    # Append the extracted data to the list
                    data.append([adminUnitName, partyName, partyCode, forename, surname, candidateRegNumber, name, value])
    return data

# Write the extracted data to a CSV file
with open('votes.csv', 'w', newline='') as file:
    writer = csv.writer(file)
    header = ['AdminUnitName', 'PartyName', 'PartyCode', 'Forename', 'Surname', 'CandidateRegNumber', 'Name', 'Value']
    writer.writerow(header)

    # kõik KOV-id
    with open('vallad.csv', 'r') as f:
        reader = csv.reader(f)
        next(reader) # Skip the header row
        for row in reader:
            parish_code = row[1]
            print(f'loen {parish_code}.xml ...')
            data = loadfile(f'{parish_code}.xml')
            if data:
                writer.writerows(data)
                
    # tallinn eraldi
    for filename in os.listdir('.'):
        if filename.startswith('VOTING_RESULT_IN_PARISH') and filename.endswith('.xml'):
            print(f'loen {filename} ...')
            data = loadfile(f'{filename}')
            if data:
                writer.writerows(data)
```

```python
# candidates from https://opendata.valimised.ee/api/RK_2023/RESULTS.xml
url = f'https://opendata.valimised.ee/api/RK_2023/RESULTS.xml'
print(url)
response = requests.get(url)
xml_content = response.content
root = ElementTree.fromstring(xml_content)

# Extract candidate data from the XML data
candidates = []
for candidate in root.iter('{https://opendata.valimised.ee/schemas/election-result/rk/v1/}candidate'):
    candidate_data = {
        'forename': candidate.find('{https://opendata.valimised.ee/schemas/election-result/rk/v1/}forename').text,
        'surname': candidate.find('{https://opendata.valimised.ee/schemas/election-result/rk/v1/}surname').text,
        'party_name': candidate.find('{https://opendata.valimised.ee/schemas/election-result/rk/v1/}partyName').text,
        'candidate_reg_number': candidate.find('{https://opendata.valimised.ee/schemas/election-result/rk/v1/}candidateRegNumber').text,
        'candidate_id': candidate.find('{https://opendata.valimised.ee/schemas/election-result/rk/v1/}candidateId').text,
        'candidate_votes': candidate.find('{https://opendata.valimised.ee/schemas/election-result/rk/v1/}candidateVotes').text,
        'district_number': candidate.find('{https://opendata.valimised.ee/schemas/election-result/rk/v1/}districtNumber').text,
        'quota': candidate.find('{https://opendata.valimised.ee/schemas/election-result/rk/v1/}quota').text,
        'vrd': candidate.find('{https://opendata.valimised.ee/schemas/election-result/rk/v1/}vrd').text,
        'v1': candidate.find('{https://opendata.valimised.ee/schemas/election-result/rk/v1/}v1').text,
        'v2': candidate.find('{https://opendata.valimised.ee/schemas/election-result/rk/v1/}v2').text,
        'v3': candidate.find('{https://opendata.valimised.ee/schemas/election-result/rk/v1/}v3').text,
        'elected': candidate.find('{https://opendata.valimised.ee/schemas/election-result/rk/v1/}elected').text,
        'reserved': candidate.find('{https://opendata.valimised.ee/schemas/election-result/rk/v1/}reserved').text
    }
    candidates.append(candidate_data)

# Write the candidate data to a CSV file
with open('candidates.csv', 'w', newline='') as csvfile:
    fieldnames = ['forename', 'surname', 'party_name', 'candidate_reg_number', 'candidate_id', 'candidate_votes', 'district_number', 'quota', 'vrd', 'v1', 'v2', 'v3', 'elected', 'reserved']
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    writer.writeheader()
    for candidate in candidates:
        writer.writerow(candidate)
```
